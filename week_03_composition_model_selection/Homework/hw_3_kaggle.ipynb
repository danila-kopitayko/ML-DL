```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
lenc = LabelEncoder()
for col in cat_cols:
  data[col] = lenc.fit_transform(data[col])
```

```python
enc = OneHotEncoder(sparse_output=False)

encoded = enc.fit_transform(data[cat_cols])
```

```python
encoded.shape
```

```python
one_hot_df = pd.DataFrame(encoded, columns=enc.get_feature_names_out(cat_cols))
```

```python
data_encoded = pd.concat([data[num_cols].reset_index(drop=True), one_hot_df.reset_index(drop=True)], axis=1)
```

```python
scaler = StandardScaler()
clf = LogisticRegression()

pipeline =  make_pipeline(scaler,clf)

param_grid=['logreg_C',[100,10,1,0.1,0.01,0.001]]
```

```python
X=data_encoded.values
y=data['Churn'].values
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(pipeline, X, y)
```
